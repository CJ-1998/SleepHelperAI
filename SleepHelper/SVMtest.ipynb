```python
import numpy as np
import nltk
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

data = pd.read_csv("/gdrive/MyDrive/Colab/SleepHelper/SampleData.csv")
# data = pd.read_csv("/gdrive/MyDrive/Colab/SleepHelper/noRateData.csv")
# data = pd.read_csv("/gdrive/MyDrive/Colab/SleepHelper/noHabitData.csv")
# data = pd.read_csv("/gdrive/MyDrive/Colab/SleepHelper/noWakeData.csv")

data.columns
data.count()

#데이터 x, y로 분리. 수면의 질 = Q가 y 데이터
x_data, y_data = data.drop(['Q'], axis=1), data['Q']

#x 데이터 정규화
x_data = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data)).values

# 데이터 train, test 데이터로 분리 
number_of_train = int(len(x_data)*0.8)

train_x = x_data[:number_of_train]
train_y = y_data[:number_of_train]
test_x = x_data[number_of_train:]
test_y = y_data[number_of_train:]
```

```python
# SVM 모델로 학습. 여기서 C를 변화시키는 것이 중요 요인이였음
svm = SVC(kernel='linear', C=1)
svm.fit(train_x, train_y) 
```

```python
#test 데이터로 정확도 측정
predict = svm.predict(test_x)

print("Accuracy : {0:f}".format(accuracy_score(predict, test_y)))
```
