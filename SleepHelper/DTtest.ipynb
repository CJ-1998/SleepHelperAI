```python
x_data, y_data = datas.drop(['Q'], axis=1), datas['Q']

# x_data = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data)).values

number_of_train = int(len(x_data)*0.8)

train_x = x_data[:number_of_train]
train_y = y_data[:number_of_train]
test_x = x_data[number_of_train:]
test_y = y_data[number_of_train:]

# Decision tree 모델 학습 ()
decision_tree = tree.DecisionTreeClassifier(criterion = 'entropy')
train_result = decision_tree.fit(train_x, train_y)

# 학습한 모델을 사용하여 예측
predict_result = decision_tree.predict(test_x)

# 예측 결과 출력 (실제 정답을 맞춘 경우 True로 표시됨)
print(predict_result == test_y)

print("Accuracy : {0:f}".format(accuracy_score(predict_result, test_y)))
```
