```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv("/gdrive/MyDrive/Colab/SleepHelper/dataSample.csv")
# data = pd.read_csv("/gdrive/MyDrive/Colab/SleepHelper/SampleData.csv")
# data = pd.read_csv("/gdrive/MyDrive/Colab/SleepHelper/noRateData.csv")
# data = pd.read_csv("/gdrive/MyDrive/Colab/SleepHelper/noHabitData.csv")
# data = pd.read_csv("/gdrive/MyDrive/Colab/SleepHelper/noWakeData.csv")
data.columns
data.count()

# data = (data - np.min(data)) / (np.max(data) - np.min(data)).values

dataset = data.values
x = dataset[:,0:13]
y = dataset[:,13]

# x = (x - np.min(x)) / (np.max(x) - np.min(x)).values

# y=tf.keras.utils.to_categorical(y)

number_of_train = int(len(dataset)*0.8)

train_x = x[:number_of_train]
train_y = y[:number_of_train]
test_x = x[number_of_train:]
test_y = y[number_of_train:]

train_y=tf.keras.utils.to_categorical(train_y)
test_y=tf.keras.utils.to_categorical(test_y)
```

```python
# 모델의 설정
model = Sequential()
model.add(Dense(30,  input_dim=13, activation='relu'))
# model.add(Dense(15, activation='relu'))
# model.add(Dense(6, activation='relu'))
model.add(Dense(6, activation='sigmoid'))

# 모델 컴파일
model.compile(loss='categorical_crossentropy',
            optimizer='adam',
            metrics=['accuracy'])

# 모델 실행
model.fit(train_x, train_y, epochs=1000, batch_size=100)

# 결과 출력
print("\n Accuracy: %.4f" % (model.evaluate(test_x, test_y)[1]))
```
