```python
import pandas as pd # Python에서 테이블화 된 데이터를 다루는데 가장 최적화 된 라이브러리입니다.
import numpy as np # numpy는 행렬이나 대규모 다차원 배열을 쉽게 처리할 수 있도록 하는 라이브러리입니다.
import matplotlib.pyplot as plt
import seaborn as sns # matplotlib을 기반으로 한 Visualization libarary 입니다.
```

```python
data = pd.read_csv("/gdrive/MyDrive/Colab/SleepHelper/SampleData.csv")
data.columns
data.count()
```

```python
print(data.info())
print()
print(data.describe())
print()
```

```python
print(data[['TST','Q']].groupby(['TST'],as_index=False).mean().sort_values(by='TST',ascending=True))
```

```python
plt.figure(figsize=(12,12))

sns.heatmap(data.corr(), linewidths=0.1, vmax=0.5, cmap=plt.cm.gist_heat, linecolor='white', annot=True)

plt.show()
```

```python
sns.pairplot(data,hue='Q')

plt.show()
```
