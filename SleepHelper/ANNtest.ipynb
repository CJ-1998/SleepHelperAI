```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import sys

from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import LabelEncoder

# data = pd.read_csv("/gdrive/MyDrive/Colab/SleepHelper/noWakeData1.csv")
data = pd.read_csv("/gdrive/MyDrive/Colab/SleepHelper/noWakeData.csv")
data.columns
data.count()

data = (data - np.min(data)) / (np.max(data) - np.min(data)).values

dataset = data.values
x = dataset[:,0:11]
y = dataset[:,11]

# x = (x - np.min(x)) / (np.max(x) - np.min(x))
# y = (y - np.min(y)) / (np.max(y) - np.min(y))

y=tf.keras.utils.to_categorical(y)
# np.set_printoptions(threshold=sys.maxsize)
# print(y)

# i=0
# for t in y:
#   i=i+1
#   if(t[0]<t[1]):
#     print(t,i)
```

```python

# 모델의 설정
model = Sequential()
model.add(Dense(30,  input_dim=11, activation='relu'))
model.add(Dense(2, activation='sigmoid'))

# 모델 컴파일
model.compile(loss='categorical_crossentropy',
            optimizer='adam',
            metrics=['accuracy'])

# 모델 실행
model.fit(x, y, epochs=10, batch_size=10)

test=[]

# for i in range(1,1440):
#   test.append([i+60,i,(i+60/i)*100,30,35,30,0,1,0,0,3])
#   test=np.array(test)
#   ans=model.predict(test)
#   test=test.tolist()
#   print(test)
#   print(ans)
#   if(ans[0][1]==1.0):
#     print(i)
#   else:
#     test=[]
#     ans=[]

# test.append([800,420,(800/420)*100,90,120,30,1,1,1,1,2])
# test.append([510,420,(510/420)*100,30,30,30,0,1,0,0,5])
# test.append([510,210,(510/210)*100,30,180,30,0,1,0,0,3])
test.append([510,360,(510/360)*100,60,60,30,0,1,0,0,5])

test=np.array(test)

ans=model.predict(test)

print(ans)

# model.save('/gdrive/MyDrive/Colab/SleepHelper/SleepHelper.h5')


# 결과 출력
# print("\n Accuracy: %.4f" % (model.evaluate(test_x, test_y)[1]))
```
