```python
import numpy as np
import nltk
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# data = pd.read_csv("/gdrive/MyDrive/Colab/SleepHelper/SampleData.csv")
# data = pd.read_csv("/gdrive/MyDrive/Colab/SleepHelper/noRateData.csv")
# data = pd.read_csv("/gdrive/MyDrive/Colab/SleepHelper/noHabitData.csv")
data = pd.read_csv("/gdrive/MyDrive/Colab/SleepHelper/noWakeData.csv")


data.columns
data.count()

x_data, y_data = data.drop(['Q'], axis=1), data['Q']

x_data = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data)).values

number_of_train = int(len(x_data)*0.8)

train_x = x_data[:number_of_train]
train_y = y_data[:number_of_train]
test_x = x_data[number_of_train:]
test_y = y_data[number_of_train:]

x_data1, y_data1 = data.drop(['Q'], axis=1), data['Q']

train_x1 = x_data1[:number_of_train]
train_y1 = y_data1[:number_of_train]
test_x1 = x_data1[number_of_train:]
test_y1 = y_data1[number_of_train:]
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_x, train_y)
print("Test Accuracy of Logistic Regression {:.2f}%".format(lr.score(test_x, test_y) * 100))

from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
# nb.fit(train_x, train_y)
nb.fit(train_x1, train_y1)
print("Accuracy of Naive Bayes: {:.2f}%".format(nb.score(test_x1, test_y1)*100))

# Random Forest Classification
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 1000, random_state = 1)
rf.fit(train_x, train_y)
print("Random Forest Algorithm Accuracy Score : {:.2f}%".format(rf.score(test_x, test_y)*100))

rf1 = RandomForestClassifier(n_estimators = 1000, random_state = 1)
rf1.fit(train_x1, train_y1)
print("Random Forest Algorithm Accuracy Score : {:.2f}%".format(rf1.score(test_x1, test_y1)*100))
```
